# __Euro Exchange Rate__

In this data project we aim to visually explore a dataset containing the exchange rates for the Euro (€) against other currencies. The dataset describes the daily Euro exchange rates between 1999 and 2021. The [Euro]( https://ec.europa.eu/info/business-economy-euro/euro-area_en) is the official currency of 19 of the 27 countries in the European Union and is used by 343 million people. 

The [Euro]( https://www.bis.org/publ/rpfx13fx.pdf) is both the second largest behind Pound Sterling (£) and the second most traded currency on the international markets after United States Dollar (UD$). 

The data set can be found on [Kaggle]( https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020), it was collected by a user called [Daria Chemkaeva]( https://www.kaggle.com/lsind18) from the European Central Bank. The dataset is often updated, so this analysis may need to be updated in the future.

#  __Data cleaning__

The next part of our project will be to clean our data set and then extract the data that are of interest for us. As stated in the introduction, we will be focusing on exchange rates for the `US dollar`, `UK pound sterling`, `Swedish krona`. We will create a copy containing our currencies of interest and the date column. 

We will start our data cleaning by renaming our columns using the pandas [`rename()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html)  function to change the names of our columns of interest. We will convert our currencies into snake_case and rename ` Period\Unit:` column to `Time`.  After we will use the pandas [`to_datetime`]( https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html?highlight=to_datetime#pandas.to_datetime) function to easier manipulate the `Time` column. 

```python
exchange_rates.rename(columns={r'Period\Unit:':'Time', '[UK pound sterling ]': 'UK_pound',
                               '[Swedish krona ]': 'SWE_krona','[US dollar ]': 'US_dollar'},
                                 inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

exchange_rates[['Time', 'UK_pound', 'SWE_krona', 'US_dollar']].head()
```

Now that we have renamed our columns and converted our Time column into a datetime object, we can begin sorting our data. We will use the pandas [`sort_values`]( https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html?highlight=sort_values#pandas.DataFrame.sort_values) to allows us to sort our data frame by our `Time` column in ascending order. We run into an issue,  as the `index` has flipped with index `5698` being our first index. To fix this we will use the pandas [`reset_index()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html?highlight=reset_index#pandas.DataFrame.reset_index) function to reset our `index` so `0` appears first. 

```python
exchange_rates.sort_values('Time', inplace = True)
exchange_rates.reset_index(drop=True, inplace= True)

exchange_rates[['Time', 'UK_pound', 'SWE_krona', 'US_dollar']].head()
```

We will now use the pandas [`copy()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.copy.html?highlight=copy#pandas.DataFrame.copy) function to copy our desired columns into a new data frame that we will use in our analysis.  From there we will look at our new column in our copy and use the pandas [`value_counts()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html?highlight=value_counts#pandas.Series.value_counts) to look at each currency to see if there are any odd values. 

```python
eruo_ex_of_intrest = exchange_rates[['Time', 'UK_pound', 'SWE_krona', 'US_dollar']].copy()

print('UK pound sterling value_count')
print(eruo_ex_of_intrest['UK_pound'].value_counts().head())
print('\n')
print('Swedish krona value_count')
print(eruo_ex_of_intrest['SWE_krona'].value_counts().head())
print('\n')
print('US dollar value_count')
print(eruo_ex_of_intrest['US_dollar'].value_counts().head())
```

We can see that there are 62 times where only a hyphen (-) appears. We can inspect these hyphen datapoints by using Boolean indexing to only select these datapoints. 

```python
eruo_hyphen = eruo_ex_of_intrest[eruo_ex_of_intrest['US_dollar'] == '-']

eruo_hyphen
```

We can see that there are 62 times where only a hyphen (-) appears. We can inspect these hyphen datapoints by using Boolean indexing to only select these datapoints. This also explains why the series was an object instead of a float. A hyphen cannot be stored as a float64, this means we need remove the hyphens so we can convert our series into a numeric data type. 

To remove the datapoints that are only hyphens, we will use `Boolean Indexing` to select all values that are not equal to `’-‘`. After that, we use the pandas [`astype`]( https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html?highlight=astype#pandas.DataFrame.astype) to convert all of our currencies from objects to floats. 

```python
euro_bool = eruo_ex_of_intrest[eruo_ex_of_intrest['UK_pound'] != '-']
euro_exchange = euro_bool.copy()

euro_exchange[['UK_pound', 'SWE_krona', 'US_dollar']] = euro_bool[['UK_pound', 'SWE_krona', 'US_dollar']].astype(float)
euro_exchange.head()
```

# __Data Visualization__

The next step for our project will be to start our data visualization. We will begin by using the [`matplotlib.pyplot.plot`]( https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html) function to plot out the exchange rates for every currency. We will also be using the [`matplotlib.pyplot.subplot`]( https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.subplot.html) function to display all of our graphs together. Finally we will use the pandas [`zip()`](https://docs.python.org/3/library/functions.html#zip) to iterate over our variables and make our graphs at the same time. 

```python
%matplotlib inline

title = ['UK pound sterling', 'Swedish krona', 'US dollar']

plt.figure(figsize=(10,10))

for i, j in zip(range(1,4), title):
    
    plt.subplot(3,1,i)
    plt.plot(euro_exchange['Time'], euro_exchange.iloc[:,i])
    plt.ylabel('Exchange Rates')
    plt.xlabel('Year')
    plt.title('Euro Exchange Against The '+ j)

plt.tight_layout()
plt.show()
```

We can see that on our 3 line graphs have several “wiggles”. This is due to daily fluctuations in the exchange rate which makes the graph much more difficult to read and analysis. We can only get an idea of upward and downward trends over the space of several months and years. 

To remove these wiggles we will use the pandas [`rolling()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.rolling.html) function to calculate the rolling mean and graph that instead, doing this will average out the extreme fluctuations. We will need to select the size of our rolling window by choosing the number of days we wan to measure the average by. We will look to see what number fits and makes our graph more readable. We will graph 7, 30, 50, 100, and 365 long windows and compare them using ` matplotlib.pyplot.subplot` function again.


```python
def rolling_mean_subplot(time, currency, title):
    rolling_mean_value = [7,30,50,100,365]
    subplot_locations = list(range(2,7))
   
    plt.figure(figsize=(9,6))
    plt.subplot(3,2,1)
    plt.plot(time, currency)
    plt.title('Original Values', weight='bold')
    plt.ylabel('Exchange Rate')
    plt.xlabel('Year')
    
    for i, j in zip(subplot_locations,rolling_mean_value):

        plt.subplot(3,2,i)
        plt.plot(time, currency.rolling(j).mean())
        plt.title('Rolling Window: ' + str(j), weight = 'bold')
        plt.ylabel('Exchange Rate')
        plt.xlabel('Year')

    plt.suptitle(title, weight = 'bold')
    plt.tight_layout()    
    plt.show()
```

```python
# UK pound Rolling Mean

rolling_mean_subplot(euro_exchange['Time'], euro_exchange['UK_pound'], 'UK Pound')

# SWE krona rolling mean

rolling_mean_subplot(euro_exchange['Time'], euro_exchange['SWE_krona'], 'SWE Krona')


# Us dollar rolling mean

rolling_mean_subplot(euro_exchange['Time'], euro_exchange['US_dollar'], 'US Dollar')

```

Now that we have graphed our rolling means for our currencies. We can see that the 30 day window is smother and gives us a greater representation of the fluctuating exchange rates, but is not to smooth where the finer detail has been removed. 

We will now add the 30 day rolling average to our `euro_exchange` data frame. We will add the columns with the pandas [`insert()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html) function. 

```python
col_names = ['UK_Pound_Rolling_Mean', 'SEK_Krona_Rolling_Mean', 'US_Dollar_Rolling_Mean']
index_num = [2,4,6]
curencies = ['UK_pound', 'SWE_krona', 'US_dollar']

for name, ind, cur in zip(col_names, index_num, curencies):
    euro_exchange.insert(ind, name, euro_exchange[cur].rolling(30).mean())
    
euro_exchange
```

## __Pre VS Post Brexit__

The first step in our Brexit analysis will be to create two copies of the data frame, one with our entire data set and one with only the post Brexit data. To do this we will use the pandas [`copy()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.copy.html?highlight=copy#pandas.DataFrame.copy) to make our copies. We do this so we can graph both data frames on top of each other to make the difference visually distinct.  

```python
brexit = euro_exchange.copy()[euro_exchange['Time'].dt.year >=2012]
post_brexit = euro_exchange.copy()[euro_exchange['Time'].dt.year >=2016]
```

We will now graph our Brexit data frames using the [`matplotlib.pyplot.plot`]( https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html) to make line graphs. The other function we will use to construct our graph include: [`matplotlib.style`]( https://matplotlib.org/3.5.1/gallery/style_sheets/style_sheets_reference.html), [`matplotlib.axvline`]( https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.axvline.html), [`matplotlib.set_xticklables`]( https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html) and [`matplotlib.text`]( https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.text.html).

```python
import matplotlib.style as style
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(8,3))


# Adding the plot
ax.plot(brexit['Time'],
        brexit['UK_Pound_Rolling_Mean'],
        linewidth = 1, color = 'grey')

# highlighting the post brexit period
ax.plot(post_brexit['Time'],
        post_brexit['UK_Pound_Rolling_Mean'], 
        linewidth = 2, color = 'red')

# Adding a line when the referendum occured
ax.axvline(x = 16800, 
          linewidth = 1,
          color = 'blue')

# removing and adding custom tick lables 
ax.set_xticklabels([])
ax.set_yticklabels([])


years = ['2012', '2014', '2016', '2018', '2020']

x= 15227.5
for year in years:
    ax.text(x, 0.675, year, alpha = 0.5, fontsize = 12)
    x+= 733
    
rates = ['0.7', '0.8', '0.9']

y= 0.695

for r in rates:
    ax.text(14950, y, r, alpha = 0.5, fontsize = 12)
    y += 0.1

# adding title and subtitle
ax.text(15000, 0.950, 'Euro-GBP Exchange Rose After 2016 UK EU Membership Referendum', weight='bold')
ax.text(15000, 0.925, 'Euro-GBP Exchange Rates between 2012 to 2021', size = 12)

# adding a signature
ax.text(14950, 0.63, 'Carl-William Palm' + ' ' * 45 + 'Source: European Central Bank', color = 'white', backgroundcolor = 'grey')


plt.show()
```
